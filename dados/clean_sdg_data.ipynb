```python
import pandas as pd

df_metadata = pd.read_excel("development_goals_metadata.xlsx", sheet_name=1)

df_metadata = df_metadata.filter(
    [
        "Table Name",
        "Region",
        "Income Group",
    ]
)

df_metadata.columns = ["pais", "regiao", "grupo_renda"]

df_metadata.head(3)
```

```python
df = pd.read_csv("development_goals.csv")

df.drop(
    columns=["Time Code"],
    inplace=True,
)

# Remove r"\s\[.+\]" from col names
df.columns = df.columns.str.replace(r"\s\[.+\]", "", regex=True)

df.columns = [
    "ano",
    "pais",
    "codigo_pais",
    "mortes_por_hiv",
    "criancas_orfas_hiv",
    "febre_tratamento_malaria",
    "emissao_co2_per_capita",
    "uso_energia_por_milhar_pib",
    "taxa_fertilidade_por_mulher",
    "percent_area_ocupada_floresta",
    "percent_jovens_alfabetizados",
    "percent_populacao_urbana_favelas",
    "populacao",
    "percent_empregos_vulneraveis_mulheres",
    "taxa_fertilidade_adolescentes",
]

df = df_metadata.merge(
    df,
    how="inner",
    left_on="pais",
    right_on="pais",
)

df.drop(
    columns=[
        "codigo_pais",
        "percent_populacao_urbana_favelas",
        "febre_tratamento_malaria",
    ],
    inplace=True,
)

df.head(3)
```
